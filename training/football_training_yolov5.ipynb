```python
import shutil
import os

# Define source and destination directories
source_dir = 'runs/detect/train3/weights'
drive_dir = '/content/drive/My Drive/YOLO_Weights'

# Ensure destination directory exists
os.makedirs(drive_dir, exist_ok=True)

# Define the file names
best_weights = 'best.pt'
last_weights = 'last.pt'

# Define the source and destination paths
best_weights_src = os.path.join(source_dir, best_weights)
last_weights_src = os.path.join(source_dir, last_weights)
best_weights_dst = os.path.join(drive_dir, best_weights)
last_weights_dst = os.path.join(drive_dir, last_weights)

# Copy the files to Google Drive
shutil.copy(best_weights_src, best_weights_dst)
shutil.copy(last_weights_src, last_weights_dst)

print("Weights have been successfully copied to Google Drive.")
```
